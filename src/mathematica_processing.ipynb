```python
import numpy as np
import h5py
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import os

%matplotlib inline
```

```python
basePath = "G:\\Il mio Drive\\conquistiamo l'Universo\\Aurelio\\risoluzione_punti\\data"
file_save_path = basePath + "\\images10k_v4\\"
file_save_path_norm = basePath + "\\images10k_norm_v4\\"

if not os.path.exists(file_save_path):
    os.makedirs(file_save_path)

if not os.path.exists(file_save_path_norm):
    os.makedirs(file_save_path_norm)
```

```python
def import_hdf5(filename):
    f = h5py.File(filename, 'r')
    a_group_key = list(f.keys())[0]
    data = list(f[a_group_key])
    return np.array(data)
```

```python
#%%define scaler
scaler = MinMaxScaler(copy=True, feature_range=(-1, 1))
```

```python
data_dict = {}
nFiles=500
samplesPerFile=20

for i in np.arange(0, nFiles):
    print("{0} su {1}".format(i+1,nFiles))
    names_orig = ["sample-{0}/original".format(j) for j in np.arange(i * samplesPerFile, (i + 1) * samplesPerFile)]
    names_diff = ["sample-{0}/diffused".format(j) for j in np.arange(i * samplesPerFile, (i + 1) * samplesPerFile)]
    file_path = "D:\\Users\\Aure.DESKTOP-HPJ2LMB\\Desktop\\punti sfocati\\exportedImgsV4\\imgSample_{0}.hdf5".format(i+1)
    
    imgs_tmp = import_hdf5(file_path)
    for j in np.arange(samplesPerFile):
        data_dict[names_orig[j]] = imgs_tmp[j, 0, :, :]
        data_dict[names_diff[j]] = imgs_tmp[j, 1, :, :]
```

```python
np.savez_compressed(file_save_path+"img56x56_10k", **data_dict)
```
